```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# Get the data
eq = pd.read_csv("all_month.csv")

# Plot the data
plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots()
ax.scatter(eq["longitude"], eq["latitude"], s=12)

# Set parameter and axis values
ax.set_xticks(np.arange(-180,190,30))   # Longitude (-180, 180)
ax.set_yticks(np.arange(-90,100,15))    # Latitude  (-90, 90)
ax.axis([-185, 185, -95, 95])

# Set titles
ax.set_title("Earthquake Points by Latitude and Longitude", fontsize=15, fontweight='bold')
ax.set_ylabel("Latitude", fontsize=13, fontweight='bold')
ax.set_xlabel("Longitude", fontsize=13, fontweight='bold')
ax.tick_params(axis="both", labelsize=13) #Eksenlerdeki saylarn boyutu

```

```python
# Plot the data
ax = eq_type.plot.barh(figsize = (10,5), y='count', x = 'type', color = 'brown', xticks=np.arange(0,13000,2000), xlim=(0,12000))

# Set titles
ax.set_title("Concussion Types", fontsize=15, fontweight='bold')
ax.set_ylabel("Types", fontsize=13, fontweight='bold')
ax.set_xlabel("Quantity", fontsize=13, fontweight='bold')
ax.tick_params(axis="both", labelsize=13) #Eksenlerdeki saylarn boyutu
```
